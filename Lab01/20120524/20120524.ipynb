# 2. Collect data from a website by parsing HTML (3p)

In this section, you are going to collect data from a website simulating the sale of Pokemon. I have prepared all the needed links in a file (`pokemon.txt`), and you have to crawl data from these links. The expected output is a `dataframe` with the following fields:

+ `SKU`: ID of Pokemon
+ `Name`: Name of Pokemon
+ `Price`: Price of Pokemon
+ `InStock`: Quantity of Pokemons in stock
+ `Categories`: The category of Pokemon
+ `Tags`: Tags of Pokemon

Your mission is to complete function `collect_data` with `course_urls_file` as your input parameter. The output should look like `pokemon_example.csv` (I list out some examples so you can easily imagine your work).

````python
pokemon_example = pd.read_csv('pokemon_example.csv')
pokemon_example
````

````python
def collect_data(course_urls_file):
    # get urls from file
    f = open(course_urls_file)
    urls = f.readlines()
    # remove '\n' at the end of url
    urls = [item[:-1] for item in urls]

    # init empty lists to store the values of each attribute.
    SKU = []
    Names = []
    Prices = []
    InStocks = []
    Categories = []
    Tags = []

    for url in urls:
        #TODO
        # send a GET request to the URL
        response = requests.get(url)

        # parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        pokemon_info_soup = soup.find("div", class_="summary entry-summary")

        # find the values of each attribute
        sku =  pokemon_info_soup.find("span", class_="sku").text.strip()
        name = pokemon_info_soup.find("h1", class_="product_title").text.strip()
        price = pokemon_info_soup.find("p", class_="price").text.strip()[1:]
        in_stock = pokemon_info_soup.find("p", class_="stock").text.strip().split(" ")[0]
        categories = [category.text.strip() for category in pokemon_info_soup.find("span", class_="posted_in").find_all("a")]
        tags = [tag.text.strip() for tag in pokemon_info_soup.find("span", class_="tagged_as").find_all("a")]

        # append the values to lists
        SKU.append(sku)
        Names.append(name)
        Prices.append(price)
        InStocks.append(in_stock)
        Categories.append(', '.join(category for category in categories))
        Tags.append(', '.join(tag for tag in tags))

    # use dataframe to store created lists
    data = pd.DataFrame({"SKU": SKU,
                         "Name": Names,
                         "Price": Prices,
                         "InStock": InStocks,
                         "Categories": Categories,
                         "Tags": Tags})

    return data
````

````python
#TEST
data_pokemon = collect_data("pokemon.txt")
assert data_pokemon.shape == (755, 6)
````

````python
BASE_URL = 'http://api.worldbank.org/v2/'
COUNTRIES = ["US", "IN", "CN", "JP", "CA", "GB", "ZA"]
INDICATORS = ['SP.POP.TOTL',
             'SP.POP.TOTL.FE.IN',
             'SP.POP.TOTL.MA.IN',
             'SP.DYN.CBRT.IN',
             'SP.DYN.CDRT.IN',
             'SP.DYN.LE00.MA.IN',
             'SP.DYN.LE00.FE.IN',
             'SE.PRM.ENRR',
             'SE.TER.ENRR',
             'SE.PRM.CMPT.ZS',
             'SE.ADT.1524.LT.ZS']

#TODO (option)
# If you need other initializations
````

````python
def collect_data(countryCode, per_page, start_year, end_year):
    #TODO
    # init empty lists to store the values of each attribute.
    Total_Population = []
    Female_Population = []
    Male_Population = []
    Birth_Rate = []
    Death_Rate = []
    Male_life_expectancy = []
    Female_life_expectancy = []
    School_enrollment_primary = []
    School_enrollment_tertiary = []
    Primary_completion_rate = []
    Literacy_rate = []
    Year = []
    Country = []

    # fill data for each attribute
    for indicator in INDICATORS:
      # create url and params
      url = BASE_URL + "countries/" + countryCode + "/indicators/" + indicator
      params = {"format": "json", "per_page": per_page, "date": f"{start_year}:{end_year}"}

      # send a GET request to the URL and return JSON content from WebAPI
      response = requests.get(url, params=params)
      data = response.json()[1]

      # fill data for each year
      for item in data:
        value = item["value"]
        year = item["date"]

        if indicator == "SP.POP.TOTL":
          if value is None:
            Total_Population.append("None")
          else: Total_Population.append(value)
          Year.append(year)
          Country.append(countryCode)
        elif indicator == "SP.POP.TOTL.FE.IN":
          if value is None:
            Female_Population.append("None")
          else: Female_Population.append(value)
        elif indicator == "SP.POP.TOTL.MA.IN":
          if value is None:
            Male_Population.append("None")
          else: Male_Population.append(value)
        elif indicator == "SP.DYN.CBRT.IN":
          if value is None:
            Birth_Rate.append("None")
          else: Birth_Rate.append(value)
        elif indicator == "SP.DYN.CDRT.IN":
          if value is None:
            Death_Rate.append("None")
          else: Death_Rate.append(value)
        elif indicator == "SP.DYN.LE00.MA.IN":
          if value is None:
            Male_life_expectancy.append("None")
          else: Male_life_expectancy.append(value)
        elif indicator == "SP.DYN.LE00.FE.IN":
          if value is None:
            Female_life_expectancy.append("None")
          else: Female_life_expectancy.append(value)
        elif indicator == "SE.PRM.ENRR":
          if value is None:
            School_enrollment_primary.append("None")
          else: School_enrollment_primary.append(value)
        elif indicator == "SE.TER.ENRR":
          if value is None:
            School_enrollment_tertiary.append("None")
          else: School_enrollment_tertiary.append(value)
        elif indicator == "SE.PRM.CMPT.ZS":
          if value is None:
            Primary_completion_rate.append("None")
          else: Primary_completion_rate.append(value)
        elif indicator == "SE.ADT.1524.LT.ZS":
          if value is None:
            Literacy_rate.append("None")
          else:
            Literacy_rate.append(value)

    # use dataframe to store created lists
    data = pd.DataFrame({"Total Population": Total_Population,
                         "Female Population": Female_Population,
                         "Male Population": Male_Population,
                         "Birth Rate": Birth_Rate,
                         "Death Rate": Death_Rate,
                         "Male life expectancy": Male_life_expectancy,
                         "Female life expectancy": Female_life_expectancy,
                         "School enrollment, primary": School_enrollment_primary,
                         "School enrollment, tertiary": School_enrollment_tertiary,
                         "Primary completion rate": Primary_completion_rate,
                         "Literacy rate": Literacy_rate,
                         "Year": Year,
                         "Country": Country})
    return data
````

````python
def Generate_Countries_Dataset(countryCode_list):
    data = pd.DataFrame()
    for countryCode in countryCode_list:
        data = pd.concat([data, collect_data(countryCode = countryCode, per_page = 100, start_year = 2000, end_year = 2022)], axis=0)
    return data
````

````python
#TEST
data_countries = Generate_Countries_Dataset(COUNTRIES)
assert data_countries.shape == (161, 13)
````
